```python
lego_colors = pd.read_excel("lego_colors.xlsx", sheet_name='color_names')
all_bick_sizes = pd.read_excel("lego_colors.xlsx", sheet_name='brick_size')
all_collors = all_bick_sizes['LEGO Name'].unique()


# color_list = ['Earth Blue', 'Medium Stone Grey', 'White', 'Bright Orange', 'Dark Stone Grey',
#               'Dark Green', 'Brick Yellow', 'Bright Blue', 'Bright Red', 'Bright Yellow', 
#               'Reddish Brown', 'Black']

# color_list = ['Medium Stone Grey', 'White', 'Dark Stone Grey',
#               'Brick Yellow', 'Reddish Brown', 'Black']

color_list = ['Earth Blue', 'Medium Stone Grey', 'White', 'Bright Orange', 'Dark Stone Grey',
              'Dark Green', 'Brick Yellow', 'Bright Blue', 'Bright Yellow', 'Reddish Brown', 
              'Black', '(New) Dark Red','Medium Lavender', 'Sand Green', 'Medium Lilac',
              'Bright Reddish Violet','Light Purple', 'Medium Blue', 'Bright Yellowish Green', 
             'Flame Yellowish Orange']

# color_list = ['Earth Blue', 'Medium Stone Grey', 'White', 'Bright Orange', 'Dark Stone Grey',
#               'Dark Green', 'Brick Yellow', 'Bright Blue', 'Bright Yellow', 'Reddish Brown', 
#               'Black', 'Sand Green', 'Medium Blue', 'Bright Yellowish Green','Flame Yellowish Orange']


brick_colors = lego_colors.loc[ lego_colors['LEGO Name'].isin(all_collors) ].copy()
brick_colors.drop(columns=["Bricklink Name", 'Available'], inplace=True)
brick_colors.reset_index(drop=True, inplace=True)

df_centroid_list = brick_colors[['Red', 'Green', 'Blue']].copy()

km = KMeans(n_clusters=len(brick_colors), random_state=0)
km.fit(df_centroid_list)
brick_colors['cluster'] = km.labels_

brick_rgb_tuple = [tuple(x) for x in brick_colors[['Red', 'Green', 'Blue']].to_numpy()]


for j in range(2):
    start = 0
    stop = plt_len = int(len(brick_rgb_tuple) / 2 + len(brick_rgb_tuple) % 2)
    if j == 1:
        start = stop
        plt_len = len(brick_rgb_tuple) - stop
        stop = len(brick_rgb_tuple)
    fig, ax = plt.subplots(1, plt_len, figsize=(30, 5))
    for i in range(start, stop):
        ax[i - start].imshow([[brick_rgb_tuple[i]]])
        ax[i - start].title.set_text(brick_colors['cluster'][i])
        ax[i - start].text(-0.2, 0.6, brick_colors['LEGO Name'][i])
        ax[i - start].axes.get_xaxis().set_visible(False)
        ax[i - start].axes.get_yaxis().set_visible(False)
```

```python
all_bick_sizes = pd.read_excel("lego_colors.xlsx", sheet_name='brick_size')
all_bick_sizes = all_bick_sizes[all_bick_sizes['Status'] == "Available"]
all_bick_sizes = all_bick_sizes.merge(brick_colors, how='left', on='LEGO Name', sort=False)
all_bick_sizes = all_bick_sizes[['Length', 'Hight', 'cluster', 'Price (CHF)', 'LEGO Name']]

all_bick_sizes['Value'] = all_bick_sizes.apply(lambda row: row['Length']*row['Hight'], axis=1)
all_bick_sizes.sort_values(by='Value', ascending=False, inplace=True)
```

```python
def overlay_effect(color, overlay):
    '''Actual overlay effect function'''
    if color < 33:
        return overlay - 100
    elif color > 233:
        return overlay + 100
    else:
        return overlay - 133 + color
    
def apply_color_overlay(image, color):
    '''Small function to apply an effect over an entire image'''
    overlay_red, overlay_green, overlay_blue = color
    channels = image.split()

    r = channels[0].point(lambda color: overlay_effect(color, overlay_red))
    g = channels[1].point(lambda color: overlay_effect(color, overlay_green))
    b = channels[2].point(lambda color: overlay_effect(color, overlay_blue))


    channels[0].paste(r)
    channels[1].paste(g)
    channels[2].paste(b)

    return Image.merge(image.mode, channels)

def make_lego_image(thumbnail_image, brick_path):
#     '''Create a lego version of an image from an image'''
    brick_image = Image.open(brick_path)
    
    base_width, base_height = thumbnail_image.shape[0:2]
    brick_width, brick_height = brick_image.size

#     rgb_image = thumbnail_image.convert('RGB')

    lego_image = Image.new("RGB", (base_width * brick_width,
                                   base_height * brick_height), "white")
    for brick_x in range(base_width):
        for brick_y in range(base_height):
            color = thumbnail_image[brick_y][brick_x]
            lego_image.paste(apply_color_overlay(brick_image, color),
                             (brick_x * brick_width, brick_y * brick_height))
    return lego_image

```

```python
#FIX RATIO --> Should be a ratio of the size defined not only a square

def img_to_thumbnail(image):
    #We need to crop the image to a square.
    #We crop the square by centering in the middle of the image
    #We select the smalles of the hight or lenght to choose how we are going to crop
    min_size = min(image.shape[0:2])
    
    #Starting points for the croping
    y_0 = math.floor((image.shape[0] - min_size) / 2)
    x_0 = math.floor((image.shape[1] - min_size) / 2)
    
    #We crop the image starting from the starting points until the min_size plus the starting point distance
    cropped_image = image[y_0: (min_size + y_0), x_0: (min_size + x_0)]
    
    resized_image = cv2.resize(cropped_image, (plate_size, plate_size), interpolation = cv2.INTER_AREA)
    
    rgb_resized_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
    
    return rgb_resized_image


def legofy(image_path, brick_colors):

    image = cv2.imread(image_path)
    
    rgb_resized_image = img_to_thumbnail(image)
    
    
    # I create the classifier and I train it
    knn = KNeighborsClassifier(n_neighbors=1, weights='distance', algorithm='auto')
    knn.fit(brick_colors[['Red', 'Green', 'Blue']],brick_colors['cluster'])

    X = rgb_resized_image.reshape(-1, 3)
    # I feed the file to get the classification of each pixel
    prediction = knn.predict(X)
#     I transform the prediction results in a dataframe
    df_prediction = pd.DataFrame(prediction, columns=['cluster'])

#     Join the prediction results with brick_colors to get the RBG of the resulting values
    new_image_cluster = df_prediction.merge(brick_colors, how='left', on='cluster', sort=False)
    new_image = new_image_cluster.drop(['cluster', 'LEGO Name'],axis=1) #Remove unessary columns
    
    # From the df I get the values
    clusterized = new_image.values

    # I take the image values and I reshape it to the original shape with RGB channels
    lego_img_no_bricks = clusterized.reshape(rgb_resized_image.shape)
    
#     Saving the image 
    lego_img_no_bricks_save = lego_img_no_bricks.astype('uint8')
    #We make the output image bigger for better visibility
    image_scaler = 20
    lego_img_no_bricks_save = cv2.resize(lego_img_no_bricks_save, (plate_size * image_scaler, plate_size * image_scaler), interpolation = cv2.INTER_NEAREST)
    cv2.imwrite(os.path.join(os.path.abspath(''), "exports", "thumnail_" + image_path), cv2.cvtColor(lego_img_no_bricks_save, cv2.COLOR_RGB2BGR))
    
#   Add lego bricks to the output image
    brick_path = os.path.join(os.path.abspath(''), "brick", "1x1.png")
    lego_image = make_lego_image(lego_img_no_bricks, brick_path)
    
    return [rgb_resized_image, lego_img_no_bricks, new_image_cluster, lego_image]
```

```python
def find_lego_bricks(image_path, brick_colors, all_bick_sizes):

    thumbnail, lego_img_no_bricks, df_lego_img, lego_image = legofy(image_path, brick_colors)

    #I get from each pixel (position on the lego plate) and the cluster (the corresponding color)
    #To do so I get for each cluster value from the resulting dataframe and reshape it to the plate_size (16x16, 32x32, 48x48, etc)
    plate_briks = df_lego_img[['cluster']].values.reshape((plate_size,plate_size))
    clusters = brick_colors['cluster'].tolist()

    #Create the output DF
    output_bricks = pd.DataFrame(columns=['x', 'y', 'Length', 'Hight','Rotated',  'cluster', 'Price', 'LEGO Name'])

    #For each color / cluster we get the local maxima --> We try to fit the biggest lego possible in that position
    for cluster in clusters:
        #start position for each color (0,0)
        row_i = column_i = 0

        #we iterate vertically on the lego plate (matrix)
        while row_i < len(plate_briks):
            #get row from row_i pointer
            row = plate_briks[row_i]

            #we iterate horizontally on the lego plate (matrix)
            while column_i < len(row):
                #get column from column_i pointer
                column = row[column_i]

                #check if the position we are in the matrix (row_i, column_i) is the same as the cluster / color we are checking
                if(column == cluster):
                    #get all available brick sizes for the given cluster / color
                    brick_sizes = all_bick_sizes.loc[all_bick_sizes['cluster'] == cluster]

                    #check if a given brick fits in the given position
                    for index, brick_size in brick_sizes.iterrows():
                        #get the length and high of the brick to check
                        l, h = brick_size[['Length','Hight']].astype('int32')
                        #check the fit horizontally then vertically
                        check_fit_h = plate_briks[row_i:(h + row_i), column_i:(l + column_i)] == cluster 
                        check_fit_v = plate_briks[row_i:(l + row_i), column_i:(h + column_i)] == cluster

                        #if the brick fits then save the position and set the pixes in the lego plate to -1 to avoid checking agian for brick fit.
                        if(check_fit_h.size == brick_size['Value'] and check_fit_h.all()):
                            output_bricks = output_bricks.append({'x': column_i, 'y':row_i, 'Length':l, 'Hight':h , 'Rotated':0, 'cluster':cluster, 'Price': brick_size['Price (CHF)'], 'LEGO Name': brick_size['LEGO Name']}, ignore_index=True)
                            plate_briks[row_i:(h+row_i), column_i:(l+column_i)].fill(-1)
                            break
                        if(check_fit_v.size == brick_size['Value'] and check_fit_v.all()):
                            output_bricks = output_bricks.append({'x': column_i, 'y':row_i, 'Length':h, 'Hight':l, 'Rotated':1, 'cluster':cluster, 'Price': brick_size['Price (CHF)'], 'LEGO Name': brick_size['LEGO Name']}, ignore_index=True)
                            plate_briks[row_i:(l+row_i), column_i:(h+column_i)].fill(-1)
                            break

                column_i = column_i + 1

            row_i = row_i + 1
            column_i = 0
            
    return output_bricks
    
```
